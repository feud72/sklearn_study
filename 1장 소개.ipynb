## 1.7 붓꽃의 품종 분류

### 1.7.1 iris_dataset

Toy dataset: [Link](https://scikit-learn.org/stable/datasets/index.html#toy-datasets)

iris_dataset: [Link](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris) [Link](https://scikit-learn.org/stable/datasets/index.html#iris-dataset)

### 1.7.2 train_test_split

sklearn.model_selection.train_test_split(\*arrays, \*\*options): [Link](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html#sklearn.model_selection.train_test_split)

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

### 1.7.3 pd.plotting.scatter_matrix

```python
import pandas as pd

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, alpha=0.8, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60)
```

### 1.7.4 k-NN, fit

sklearn.neighbors.KNeighborsClassifier: [Link](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier)

class sklearn.neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None, \*\*kwargs)[source](https://github.com/scikit-learn/scikit-learn/blob/7e85a6d1f/sklearn/neighbors/_classification.py#L25)

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
```

fit(self, X, y)

Fit the model using X as training data and y as target values

```python
knn.fit(X_train, y_train)
```

### 1.7.5 predict

꽃받침의 길이가 5cm, 폭이 2.9cm이고 꽃잎의 길이가 1cm, 폭이 0.2cm인 붓꽃을 보았다고 가정합시다. 이 붓꽃의 품종은 무엇일까요?

```python
X_new = np.array([[5, 2.9, 1, 0.2]])

print("X_new.shape: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)

print("예측: {}".format(prediction))

print("예측한 타깃의 이름: {}".format(iris_dataset['target_names'][prediction]))
```

### 1.7.6 score

```python
y_pred = knn.predict(X_test)

print("테스트 세트에 대한 예측값:\n {}".format(y_pred))
```

```python
print("테스트 세트의 정확도:\n {:.3f}".format(np.mean(y_pred == y_test)))
```

```python
print("테스트 세트의 정확도:\n {:.3f}".format(knn.score(X_test, y_test)))
```
